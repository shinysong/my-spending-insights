```python
import pandas as pd
```

```python
df = pd.read_csv('..\data\크롤링용.csv')
df
```

```python
category_1 = pd.read_csv('..\data\category_1.csv')
category_1
```

```python
# df의 '이미지_바로가기링크' 컬럼과 category_1의 '이미지_바로가기링크' 컬럼을 비교하여 df를 기준으로 차집합 구하기
df2 = df[~df['이미지_바로가기링크'].isin(category_1['이미지_바로가기링크'])]
df2
```

```python
df2.to_csv('..\data\크롤링용_2.csv', index=False, encoding='utf-8-sig')
```

```python
tmp = pd.read_csv('..\data\크롤링용_2.csv')
category_2 = pd.read_csv('..\data\category_2.csv')
df3 = tmp[~tmp['이미지_바로가기링크'].isin(category_2['이미지_바로가기링크'])]
df3
```

```python
df3.to_csv('..\data\크롤링용_3.csv', index=False, encoding='utf-8-sig')
```
